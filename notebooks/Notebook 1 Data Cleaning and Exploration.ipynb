```python
# This creates a full address to pass to a google maps API query
def addresscreator(df):
    df['full_address'] = df['town'] + ', ' + df['street_name'] + ' Block Number ' + df['block'] + ' Singapore'
    df['full_address'] = df.full_address.str.title()
    return df
```

```python
sellrent=pd.read_csv('../data/external/number-of-sold-and-rented-hdb-residential-units.csv')
# Restructure the table so that sold and rented counts appear side by side for each year
sr_sold = sellrent.loc[sellrent['category'] == 'Sold'].drop(columns = 'category')
sr_rent = sellrent.loc[sellrent['category'] == 'Rented'].drop(columns = 'category')
```

```python
sr_rent.rename(columns={'no_of_units' : 'hdbs_rented'}, inplace=True) 
sr_sold.rename(columns={'no_of_units' : 'hdbs_sold'}, inplace=True) 
```

```python
sr_final = sr_rent.merge(sr_sold, how='inner', on=['financial_year','property_type','flat_type'])
sr_final
```
